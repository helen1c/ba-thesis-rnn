```python
def logistic(x):
    return 1./(1 + np.exp(-x))

def tanh(x):
    return 2*logistic(2*x) - 1

def relu(x):
    return np.maximum(x, 0)

def dlogistic(x):
    return logistic(x) * (1 - logistic(x))

def dtanh(x):
    return 1 - tanh(x)**2

def drelu(x):
    return x >= 0
```
